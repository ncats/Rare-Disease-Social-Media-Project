```python
from top2vec import Top2Vec
import requests
```

```python
response = requests.get("https://github.com/ncats/Rare-Disease-Social-Media-Project/raw/main/project_data/models/ehlersdanlos/ehlersdanlos_doc2vec_eom_4")
with open("ehlersdanlos_doc2vec_eom_4", "wb") as f:
    f.write(response.content)
```

```python
EOMmodel = Top2Vec.load("ehlersdanlos_doc2vec_eom_4")
```

```python
len(EOMmodel.documents)
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
```

```python
documents, document_scores, document_ids = EOMmodel.search_documents_by_keywords(keywords=["MCAS", "mast cell"], num_docs=50)
for doc, score, doc_id in zip(documents, document_scores, document_ids):
    print(f"Document: {doc_id}, Score: {score}")
    print("-----------")
    print("\n".join([sent for sent in sent_tokenize(doc) if "MCAS" in sent]))
    print("-----------")
    print()
```
